```python
# 读取excel
excel = pd.read_excel(r"C:\Users\chenxing\Desktop\202501030113544986 .xls", sheet_name='Sheet1')
cols = ['机构', '组织', '级别','岗位', '所属', '职级', '职位代码', '招聘人数', '报考专业', '学历要求', '学位', '年龄', '经历', '其他', '申论类别', '专业科目', '备注', '咨询电话']
excel.columns = cols
excel.drop([0, 1, excel.index[-1]], inplace=True)
excel.index=np.arange(1, len(excel) + 1)
del excel['咨询电话']
excel.fillna('无', inplace=True)
excel[excel['招聘人数'] >= 5] 
```
