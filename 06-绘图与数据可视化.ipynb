```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
fig = plt.figure()
# 需要在figure创建一个或多个subplot才能绘图
ax1 = fig.add_subplot(2, 2, 1) # 表示图像是2*2，且当前选中第一张
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 1, 2)
# 绘制线型图
ax1.plot(np.random.standard_normal(50).cumsum(), color = 'black', linestyle='dashed')
# 散点图
ax2.scatter(np.arange(30), np.arange(30) + 3 * np.random.standard_normal(30))
# 直方图
ax3.hist(np.random.standard_normal(100), bins=50, color='pink', alpha = 1); # 分号可以屏蔽notebook输出最后一行的返回值
```

```python
# plt.subplots提供更便捷的方法
fig, axes = plt.subplots(2, 3)
axes
# 可以直接通过索引指定绘图表格
axes[0, 2].hist(np.random.standard_normal(30), bins=10)
# 调整子图边距
fig.subplots_adjust(wspace=0.3, hspace=0)
```

```python
# Series和DataFrame默认都有一个plot绘制
s = pd.Series(np.random.standard_normal(10).cumsum(), index=np.arange(0, 100, 10))
s.plot()
```

```python
df = pd.DataFrame(np.random.standard_normal((10, 4)).cumsum(0), columns=['A', 'B', 'C', 'D'], index=np.arange(0, 100, 10))
plt.style.use('grayscale') # 配色方案改为灰度
df.plot()
```
